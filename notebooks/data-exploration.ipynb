# Extetend Flights dataset with timestamp

```python
flights_df = pandas.read_csv('data/flights.csv')

f = flights_df
flights_df['date'] = pandas.to_datetime(flights_df[['year', 'month', 'day', 'hour', 'minute']])

flights_df.head()
```

# Count covered days

```python
flights_per_day_count_df = flights_df.groupby(flights_df['date'].dt.date).size()

number_of_groups_by_dates = flights_per_day_count_df.shape[0] # Rows count in pandas

print('Total days covered:', number_of_groups_by_dates)
```

# Count departure cities

```python
flights_per_destination_count_df = flights_df.groupby(flights_df['origin']).size().sort_index().reset_index()

flights_per_destination_count_df.columns = ['origin','count']

print(flights_per_destination_count_df)


number_of_destinations = flights_per_destination_count_df.shape[0] # Rows count in pandas

print('Total departure cities:', number_of_destinations)
```

# Exploring Flights and Planes

```python

planes_df['tailnum'] = planes_df['tailnum'].astype(str)
flights_df['tailnum'] = flights_df['tailnum'].astype(str)

flights_and_planes_df = pandas.concat([planes_df, flights_df], axis=1, sort=False)
#flights_and_planes_df.head()
#print(flights_and_planes_df.dtypes)
print(flights_and_planes_df.shape)
```

# Exploring which plane manufacturer are the most popular

```python
import numpy as np

flights_and_planes_exploration_df = flights_and_planes_df[['manufacturer', 'model', 'engines', 'dep_delay','arr_delay','distance']]

#print(df.shape)

grouped = flights_and_planes_exploration_df.groupby(['manufacturer']).size()

grouped.head(1000)
```

## Looking into most popular planes' flights

```python
df = flights_and_planes_exploration_df

most_popular_manufactorers = ['AIRBUS', 'AIRBUS INDUSTRIE', 'BOEING', 'BOMBARDIER INC', 'EMBRAER', 'MCDONNELL DOUGLAS', 'MCDONNELL DOUGLAS AIRCRAFT CO', 'MCDONNELL DOUGLAS']

flights_of_most_popular_planes = df.loc[df['manufacturer'].isin(most_popular_manufactorers)]

grouped = flights_of_most_popular_planes.groupby(['manufacturer', 'distance']).agg('min', 'max', 'avg')

grouped.head(1000)
```
