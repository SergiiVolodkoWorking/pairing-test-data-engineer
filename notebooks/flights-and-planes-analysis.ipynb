# Flight and planes analysis

## Load the data

```python
import pandas as pd

planes_df = pd.read_csv('./data/planes.csv')
flights_df = pd.read_csv('./data/flights.csv')
```

## Joining flights and planes by tail number

```python
flights_and_planes_df = pd.merge(planes_df, flights_df, on='tailnum')
#flights_and_planes_df.head()
#print(flights_and_planes_df.dtypes)
print(flights_and_planes_df.shape)
```

## Exploring which plane manufacturers are the most popular

```python
import numpy as np

df = flights_and_planes_df[['manufacturer', 'model', 'engines', 'dep_delay','arr_delay','distance']]

#print(df.shape)

flights_by_planes = df.groupby(['manufacturer']).size()

flights_by_planes.head(1000)
```

# Analyzing flight data of the most popular planes

```python
most_popular_manufactorers = ['AIRBUS', 'AIRBUS INDUSTRIE', 'BOEING', 'BOMBARDIER INC', 'EMBRAER', 'MCDONNELL DOUGLAS', 'MCDONNELL DOUGLAS AIRCRAFT CO', 'MCDONNELL DOUGLAS']

df = flights_and_planes_df

flights_of_most_popular_planes = df.loc[df['manufacturer'].isin(most_popular_manufactorers)]

flights_by_planes = flights_of_most_popular_planes.groupby(['manufacturer'])

flights_by_planes.describe()
```

## Flight distance statistics

```python
flights_by_planes['distance'].describe()
```

### Conclusions:
- Planes of `MCDONNELL DOUGLAS AIRCRAFT CO`, `MCDONNELL DOUGLAS`, `EMBRAER` and `BOMBARDIER INC` are used for flights <1600 miles
- Planes of `AIRBUS INDUSTRIE` are used for flights <2600 miles
- Planes of `AIRBUS`, `BOEING`are covering around 5000 mi flights

## Flight departure delay statistics

```python
flights_by_planes['dep_delay'].describe()
```

### Conclusions:
- About 50% of all flights departs earlier
- `BOEING` and `AIRBUS` has the most delayed departures

## Flight arrival delay statistics

```python
flights_by_planes['arr_delay'].describe()
```

### Conclusions:
- 25% of all flights arrive more then 10 minutes earlier
- Other 25% of flights arrive at least 2 minutes earlier
- `BOEING` and `AIRBUS` has the most delayed arrivals
